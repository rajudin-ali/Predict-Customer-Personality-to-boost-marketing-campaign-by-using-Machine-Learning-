## Outliers

```python
# Menghitung Q1, Q3, dan IQR
Q1 = df['Income'].quantile(0.25)
Q3 = df['Income'].quantile(0.75)
IQR = Q3 - Q1

# Mendefinisikan batas bawah dan batas atas
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Menghapus outlier
df = df[(df['Income'] >= lower_bound) & (df['Income'] <= upper_bound)].copy()

print("\nJumlah baris setelah menghapus outlier:", len(df))
```

## Data Modeling

### Elbow Method

```python
from sklearn.cluster import KMeans
X = dataset_scaled.drop(columns=['ID', 'Year_Birth', 'Education', 'Marital_Status', 'Age_Category', 'Dt_Customer', 'Income_Category', 'Spending_Category', 'Marital_Status_Grouped', 'Income_Group', 'Education_mapped']).copy()
# X = dataset_scaled[numerical_features].copy()
inertia = []

for i in range(2, 20):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X.values)
    inertia.append(kmeans.inertia_)

print(inertia)
plt.figure(figsize=(20, 10))
plt.title('Elbow Method of K-means Clustering')
# plt.plot(inertia)

sns.lineplot(x=range(2, 20), y=inertia, color='#000087', linewidth = 4)
sns.scatterplot(x=range(2, 20), y=inertia, s=300, color='#800000',  linestyle='--')
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(X.values)
X['cluster'] = kmeans.labels_
df['cluster'] = kmeans.labels_
```

### Silhouette Score

```python
from sklearn.metrics import silhouette_score

silhouette_scores = {}
k_range = range(2, 11) # Mulai dari k=2, karena Silhouette Score tidak terdefinisi untuk k=1

# Drop non-numeric columns before calculating silhouette score
dataset_scaled_numeric = dataset_scaled.drop(columns=['Education', 'Marital_Status', 'Dt_Customer', 'Income_Category', 'Spending_Category', 'Age_Category', 'Marital_Status_Grouped', 'Income_Group', 'Education_mapped'])

print("Menghitung Silhouette Score untuk k=2 hingga k=10...")
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init='auto')
    kmeans.fit(dataset_scaled_numeric)
    # Menghitung skor: rata-rata jarak antara klaster (b) vs. rata-rata jarak dalam klaster (a)
    score = silhouette_score(dataset_scaled_numeric, kmeans.labels_)
    silhouette_scores[k] = score
    print(f"k={k}: Silhouette Score = {score:.4f}")

# Menentukan k optimal
optimal_k_silhouette = max(silhouette_scores, key=silhouette_scores.get)

print(f"\nNilai k optimal berdasarkan Silhouette Score (Skor Tertinggi): {optimal_k_silhouette}")
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples
from matplotlib import cm

# Gunakan K = 2
FINAL_K = 4

# Jalankan K-Means
kmeans = KMeans(n_clusters=FINAL_K, random_state=42, n_init=10)
cluster_labels = kmeans.fit_predict(dataset_scaled_numeric) # Use dataset_scaled_numeric

# Hitung skor siluet
silhouette_avg = silhouette_score(dataset_scaled_numeric, cluster_labels) # Use dataset_scaled_numeric
sample_silhouette_values = silhouette_samples(dataset_scaled_numeric, cluster_labels) # Use dataset_scaled_numeric

# Plot silhouette
fig, ax1 = plt.subplots(1, 1, figsize=(10, 7))
y_lower = 10

for i in range(FINAL_K):
    ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
    ith_cluster_silhouette_values.sort()
    size_cluster_i = ith_cluster_silhouette_values.shape[0]
    y_upper = y_lower + size_cluster_i

    color = cm.nipy_spectral(float(i) / FINAL_K)
    ax1.fill_betweenx(np.arange(y_lower, y_upper),
                      0, ith_cluster_silhouette_values,
                      facecolor=color, edgecolor=color, alpha=0.7)

    ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
    y_lower = y_upper + 10

ax1.set_title(f"Silhouette Analysis untuk K-Means Clustering (K={FINAL_K})")
ax1.set_xlabel("Nilai Koefisien Siluet")
ax1.set_ylabel("Nomor Cluster")

ax1.axvline(x=silhouette_avg, color="red", linestyle="--",
            label=f'Skor Rata-rata: {silhouette_avg:.4f}')
ax1.set_yticks([])
ax1.legend()
plt.show()

print(f"\nSkor Siluet Rata-rata untuk K={FINAL_K}: {silhouette_avg:.4f}")
```

```python
# Cluster Graph 2D
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt


# Reduksi dimensi menjadi 2D (PCA)
pca = PCA(n_components=4)
reduced_data = pca.fit_transform(dataset_scaled_numeric)

# Plot clustering
plt.figure(figsize=(8,6))
plt.scatter(reduced_data[:, 0], reduced_data[:, 1],
            c=cluster_labels, cmap="Set1", s=50, alpha=0.7, edgecolors="k")

# Tambahkan titik centroid ke grafik
centroids_2d = pca.transform(kmeans.cluster_centers_)
plt.scatter(centroids_2d[:, 0], centroids_2d[:, 1],
            marker="X", s=200, c="yellow", edgecolor="black", label="Centroid")

plt.title("2D Cluster Graph (K=4)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.show()
```
